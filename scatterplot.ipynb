**First Visualization:** <br>
Scatter Plot: 
- x-axis: release_year 
- y-axis: date_added
- color: type (movie or TV)


imports and adding the data as the netflix dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

netflix = pd.read_csv('netflix_titles.csv')
```

creating a new column called date_added_datetime, where we are making the date_added attribute a datetime attribute, so we can use it in our visualization. 

```python
netflix['date_added_datetime'] = pd.to_datetime(netflix['date_added'], format='%B %d, %Y', errors='coerce')
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x=netflix["release_year"], y=netflix["date_added_datetime"], hue=netflix["type"], alpha=0.7)
plt.title("Date Added to Netflix vs Release Year By Type")
plt.xlabel('Release Year')
plt.ylabel('Date Added to Netflix')
plt.legend(title='Type')
plt.show()
```
