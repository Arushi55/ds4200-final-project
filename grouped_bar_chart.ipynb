```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

netflix = pd.read_csv('netflix_titles.csv')
```

```python
netflix_genres = netflix.assign(genre=netflix['listed_in'].str.split(', ')).explode('genre')
```

```python
genre_counts = netflix_genres.groupby(['genre', 'type']).size().reset_index(name='count')
```

```python
plt.figure(figsize=(16, 8))
sns.barplot(data=genre_counts, x='genre', y='count', hue='type')
plt.title('Number of Movies and TV Shows By Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=90)
plt.legend(title='Type')
plt.tight_layout()
plt.show()
```
